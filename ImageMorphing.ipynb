```python
def upload_and_read_image():
    uploaded = files.upload()
    filename = list(uploaded.keys())[0]
    image = cv2.imread(filename)
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    return image_rgb
```

```python
def resize_image(image, size):
    return cv2.resize(image, size, interpolation=cv2.INTER_AREA)
```

```python
print("Upload the first image:")
image1_rgb = upload_and_read_image()
```

```python
desired_size = (640, 480)  # Width, Height
resized_image1 = resize_image(image1_rgb, desired_size)
```

```python
print("Upload the second image:")
image2_rgb = upload_and_read_image()
```

```python
resized_image2 = resize_image(image2_rgb, desired_size)
```

```python
# Blend the two images
alpha = 0.5  # Blending factor
blended_image = cv2.addWeighted(resized_image1, alpha, resized_image2, 1 - alpha, 0)
```

```python
# Display the images
plt.figure(figsize=(15, 10))
```

```python
# @title
plt.subplot(1, 3, 1)
plt.title("First Image")
plt.imshow(resized_image1)
plt.axis('on')
```

```python
# @title
plt.subplot(1, 3, 2)
plt.title("Second Image")
plt.imshow(resized_image2)
plt.axis('off')
```

```python
# @title
plt.subplot(1, 3, 3)
plt.title("Blended Image")
plt.imshow(blended_image)
plt.axis('off')

plt.show()
```
